```python
import csv
import pandas as pd
import datetime
```

```python
coinNamesDataFrame = pd.read_csv('/home/rr/Desktop/TradingMasters/Top60.csv',header=0,dtype={'Count': int})
```

```python
def populate_tweets_df(path):
    #Declare a new data frame with pandas, with some specific column names
    tweets = pd.DataFrame(columns=['Coin', 'Time', 'Tweet', 'Retweet'])

    #Open the text file that contains the tweets we collected
    tweets_file = pd.read_csv(path, header=0,dtype={'Retweet': int})
    
    #Read the text file line by line
    for line in tweets_file.itertuples():
        tweets.loc[len(tweets)]=[line.Coin, line.Time, line.Tweet, line.Retweet]    

    return tweets

CoinTalk_tweets = populate_tweets_df('CoinTalkGeo-22.csv')
```

```python
#Remove tweets that don't include coin talk in their text
for row in range(len(CoinTalk_tweets)):
    hasCoins = 0
    for coin in coinNamesDataFrame.itertuples():
        if coin.CoinName.lower() in CoinTalk_tweets.loc[row,'Tweet'].lower():
            hasCoins += 1
        if coin.ShortName.lower() in CoinTalk_tweets.loc[row,'Tweet'].lower():
            hasCoins += 1
    
    
    CoinTalk_tweets.loc[row,'CoinCount']=hasCoins  

    if hasCoins > 1:
        CoinTalk_tweets.loc[row,'trash']=False
    else:
        CoinTalk_tweets.loc[row,'trash']=True 
    
```
